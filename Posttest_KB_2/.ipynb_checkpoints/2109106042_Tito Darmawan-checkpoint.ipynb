### Membaca Info Dataset
> <b> on_bad_line="skip" </b>: mengabaikan line yang rusak atau error dalam data set

```python
import pandas as pd
import numpy as np
from prettytable import PrettyTable

df = pd.read_csv("books.csv", on_bad_lines="skip")
df.info()
```

### Memasukan info ke dalam dictionary dan menampilkannya <br>
> <b> df.shape[index] </b>: mengambil index pertama dari df.shape = [records, atribute], yang di mana isinya jumlah records <br>
  <b> df.columns.values </b>: mengambil atribute dari setiap kolom <br>
  <b> df.select_dtypes(["tipe data"]).columns </b>: mengambil atribute dari setiap kolom dengan tipe objek <br>
  <b> df.loc[:, df.isnull().any()].columns) </b>: mengambil atribute yang memiliki record value null dari setiap kolom 

```python
dataInfo = {
    "Record"         : df.shape[0],
    "Atribute"       : list(df.columns.values),
    "Atribute Objek" : list(df.select_dtypes(["object"]).columns),
    "Atribute Angka" : list(df.select_dtypes(["int64", "float64"]).columns),
    "Atribute Null"  : list(df.loc[:, df.isnull().any()].columns),
}
keys = list(dataInfo.keys())

dataInfoTable = PrettyTable()
dataInfoTable.field_names = ["Info", "Jumlah", "Atribute"]
dataInfoTable.add_row([keys[0], dataInfo[keys[0]], "-"])
for i in range(1, 5):
    dataInfoTable.add_row([keys[i], len(dataInfo[keys[i]]), ", \n".join(dataInfo[keys[i]])])
print(dataInfoTable)
```

```python
df.corr()
```

### Mengubah atribute int64 ke float64. <br>
> <b> df["nama kolom"].astype("tipe data") </b> : mengubah tipe data

```python
df["ratings_count"] = df["ratings_count"].astype("float64")
df["ratings_count"].dtypes
```

### Menampilkan 10 record pertama atribute bertipe angka. <br>
> <b> df.head(jumlah record) </b> : menampilkan record dari atas

```python
df_numeric = df[dataInfo[keys[3]]]
df_numeric.head(10)
```

### Menyimpan hanya data yang atributenya bertipe angka. <br>
> <b> df.to_csv </b> : menyimpan data ke csv <br>
  <b> index=False </b> : Tidak mengikutkan index sebagai kolom pertama

```python
df_numeric.to_csv("books_numeric.csv", index=False)
```
